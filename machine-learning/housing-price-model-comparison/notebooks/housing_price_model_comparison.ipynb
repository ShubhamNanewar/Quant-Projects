# Housing Price Model Comparison

This notebook presents a cleaned version of the first machine learning assignment. The goal is to compare modeling housing prices in levels versus logs, using linear regression, Lasso, and Ridge under fixed holdout splits.

## Setup

The core regression idea is

$$y_i = eta_0 + x_i^\prime eta + arepsilon_i$$

and the alternative target is $\log(y_i)$. Lasso and Ridge then add $L_1$ and $L_2$ penalties to the same linear structure.

## Takeaways

- The price target performs better than the log target on this dataset.
- Ridge is the strongest of the price-level models, but its improvement over the plain linear benchmark is not statistically strong within the same target family.
- The best price model does beat the best log model clearly in prediction-error terms.

That makes the practical recommendation simple: for this dataset, model price directly and use Ridge as the main production candidate.
