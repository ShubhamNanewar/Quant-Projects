# Kalman Filter Pairs Trading

This notebook mirrors the project script and focuses on a state-space pairs-trading strategy for `RF` and `SCHW`. The trading signal comes from the Kalman filter innovation rather than a fixed-regression residual.

## State-Space Setup

The observation equation is

$$
y_t = \alpha_t + \beta_t x_t + \varepsilon_t,
$$

with latent state

$$
\theta_t = \begin{bmatrix} \alpha_t \\ \beta_t \end{bmatrix},
$$

and random-walk transition

$$
\theta_t = \theta_{t-1} + \omega_t.
$$

At each date the filter produces a one-step-ahead forecast error $e_t$ and forecast variance $F_t$. The standardized signal is

$$
z_t = \frac{e_t}{\sqrt{F_t}}.
$$

This $z$-score is the basis for trade entry and exit.
