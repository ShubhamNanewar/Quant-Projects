## Risk Model Setup

If $r_t$ is the daily return vector and $w$ is the portfolio weight vector, then portfolio return is

$$
r_{p,t} = w^\top r_t,
$$

and EUR loss is

$$
L_t = -V_0 r_{p,t}.
$$

The notebook compares empirical tail loss, normal parametric VaR/ES, and Student-t VaR/ES with degrees of freedom selected from pooled standardized residuals.
